## Airplane Crashes and Fatalities Since 1908

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Airplane_Crashes_and_Fatalities_Since_1908.csv")
```

```python
df.head()
```

```python
df2=df[['Date','Fatalities','Operator']]
```

```python
df2.head()
```

```python
# İndeksi birleştirme
df['Index'] = df['Date'] + '_' + df['Operator']
df = df.drop_duplicates(subset='Index')
df = df.drop(columns='Index')

# Tarihleri datetime formatına dönüştürme
df['Date'] = pd.to_datetime(df['Date'])

# Tarihe göre sıralama
df_sorted = df.sort_values(by='Date')

# Operatörleri sütun olarak transpose etme
df_transposed = df_sorted.pivot(index='Date', columns='Operator', values='Fatalities')

print(df_transposed)
```

```python
df_transposed.fillna(0, inplace=True)
```

```python
df_transposed.head()
```

```python
sum_daily=df_transposed.cumsum(axis=0)
sum_daily
```
